```python
#Importing Libraries
from tkinter import *
from tkinter import messagebox

def reset_entry():               #Reset Function
    age.delete(0,'end')
    height_tf.delete(0,'end')
    weight_tf.delete(0,'end')

def calculate_bmi():             #Calculate BMI Function
    kg = int(weight_tf.get())
    m = int(height_tf.get())/100
    bmi = kg/(m*m)
    bmi = round(bmi, 1)
    bmi_index(bmi)

def bmi_index(bmi):             #Weight Analysis from BMI value
    
    if bmi < 18.5:
        messagebox.showinfo('bmi-python', f'BMI = {bmi} is Underweight')
    elif (bmi > 18.5) and (bmi < 24.9):
        messagebox.showinfo('bmi-python', f'BMI = {bmi} is Normal')
    elif (bmi > 24.9) and (bmi < 29.9):
        messagebox.showinfo('bmi-python', f'BMI = {bmi} is Overweight')
    elif (bmi > 29.9):
        messagebox.showinfo('bmi-python', f'BMI = {bmi} is Obesity') 
    else:
        messagebox.showerror('bmi-python', 'something went wrong!')   

root = Tk()   #Creating GUI 
root.title('Python Miniproject D9A')
root.geometry('400x300')
root.config(bg='#113e21')

var = IntVar()

frame = Frame(
    root,
    padx=10, 
    pady=10,
    bg = '#b38b59'
)               #Making frame for accepting inputs
frame.pack(expand=True)

#Entering age - label and entry widget
age_lb = Label(
    frame,
    text="Enter Age (2 - 120)",
    bg = '#b38b59'
)
age_lb.grid(row=1, column=1)

age = Entry(
    frame, 
    bg = '#fefefe'
)
age.grid(row=1, column=2, pady=5)

#Entering gender- Label and entry widget
gen_lb = Label(
    frame,
    text='Select Gender',
    bg = '#b38b59'
)
gen_lb.grid(row=2, column=1)

frame2 = Frame(
    frame,
    bg = '#b38b59'
)
frame2.grid(row=2, column=2, pady=5)

male_rb = Radiobutton( #radiobuttons for gender (male)
    frame2,
    text = 'Male',
    variable = var,
    value = 1,
    bg = '#b38b59'
)
male_rb.pack(side=LEFT)

female_rb = Radiobutton( #radiobutton for gender (female)
    frame2,
    text = 'Female',
    variable = var,
    value = 2,
    bg = '#b38b59'
)
female_rb.pack(side=RIGHT)

#Height input - label and entry widget
height_lb = Label(
    frame,
    text="Enter Height (cm)  ",
    bg = '#b38b59'
)
height_lb.grid(row=3, column=1)

height_tf = Entry(
    frame,
    bg = '#fefefe'
)
height_tf.grid(row=3, column=2, pady=5)

#Weight input - label and entry widget
weight_lb = Label(
    frame,
    text="Enter Weight (kg)  ",
    bg = '#b38b59'
)
weight_lb.grid(row=4, column=1)

weight_tf = Entry(
    frame,
    bg = '#fefefe'
)
weight_tf.grid(row=4, column=2, pady=5)

frame3 = Frame(
    frame,
    bg = '#b38b59'
)
frame3.grid(row=5, columnspan=3, pady=10)

cal_btn = Button( #calculate button calculates the BMI from values entered
    frame3,
    text='Calculate',
    command=calculate_bmi
)
cal_btn.pack(side=LEFT)

reset_btn = Button( #reset button for reseting the values
    frame3,
    text='Reset',
    command=reset_entry
)
reset_btn.pack(side=LEFT)

exit_btn = Button( #exit button to close the GUI
    frame3,
    text='Exit',
    command=lambda:root.destroy()
)
exit_btn.pack(side=RIGHT)

root.mainloop()
```
